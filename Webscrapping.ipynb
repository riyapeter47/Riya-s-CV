```python
url="https://www.amazon.in/New-Apple-iPhone-XR-64GB/product-reviews/B08L8BK9V6/ref=cm_cr_dp_d_show_all_btm?ie=UTF8&reviewerType=all_reviews"
```

```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
page = requests.get(url)
page
```

```python
soup = BeautifulSoup(page.content, 'html.parser')
print(soup.prettify())
```

```python
names = soup.find_all('span', class_='a-profile-name')[2:]
len(names)
```

```python
ratings = soup.find_all('i', class_='review-rating')[2:]
ratings
```

```python
rating=soup.find_all("span", class_="a-icon-alt")[9:]
len(rating)
```

```python
title = soup.find_all("a", class_='a-size-base a-link-normal review-title a-color-base review-title-content a-text-bold')
len(title)
```

```python
date = soup.find_all('span', class_='review-date')[2:]
len(date)
```

```python
body= soup.find_all('span', {"data-hook":"review-body"})
len(body)
```

```python
cust_name=[]
for i in range(0,len(names)):
  b=names[i].get_text()
  cust_name.append(b)
cust_name  
```

```python
cust_ratings=[]
for i in range(0,len(ratings)):
  b=ratings[i].get_text()
  cust_ratings.append(b)
cust_ratings 
```

```python
cust_title=[]
for i in range(0,len(title)):
  b=title[i].get_text()
  cust_title.append(b)
cust_title[:]=[data.strip('\n') for data in cust_title]
cust_title  
```

```python
cust_date=[]
for i in range(0,len(date)):
  b=date[i].get_text()
  cust_date.append(b)
cust_date
```

```python
cust_body=[]
for i in range(0,len(body)):
  b=body[i].get_text()
  cust_body.append(b)
cust_body[:]=[data.strip('\n') for data in cust_body]
cust_body  
```

```python
df = pd.DataFrame()
df['date']=cust_date
df['name']= cust_name
df['rating']=cust_ratings
df['review title']= cust_title
df['review_body']= cust_body
df
```

```python

df.to_csv("amazon_reviews.csv", index= True)
```

```python
df.reset_index(inplace=True)
df['index']= df['index']+1
df
```
